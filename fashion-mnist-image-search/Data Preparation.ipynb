### Data Preparation
This notebook is being used to prepare the Fashion MNIST data for simple feedforward network.

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import tensorflow as tf
import csv

%matplotlib inline
```

```python
(x_train, y_train), (x_test, y_test) = tf.keras.datasets.fashion_mnist.load_data()
```

```python
with open('../datasets/fashion-mnist/training_data.csv', mode='w') as training_file:
    training_file = csv.writer(training_file, delimiter=',', quotechar='"', quoting=csv.QUOTE_MINIMAL)
    for item,label in zip(x_train,y_train):
        row = list(item.ravel())
        row.insert(0,label)
        training_file.writerow(row)
df = pd.read_csv("../datasets/fashion-mnist/training_data.csv",header=None)
df.head()
```

```python
with open('../datasets/fashion-mnist/test_data.csv', mode='w') as test_file:
    test_file = csv.writer(test_file, delimiter=',', quotechar='"', quoting=csv.QUOTE_MINIMAL)
    for item,label in zip(x_test,y_test):
        row = list(item.ravel())
        row.insert(0,label)
        test_file.writerow(row)
df = pd.read_csv("../datasets/fashion-mnist/test_data.csv",header=None)
df.head()
```
